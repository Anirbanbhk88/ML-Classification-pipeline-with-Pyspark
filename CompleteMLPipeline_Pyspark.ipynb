```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#create pyspark session
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('ml-pipeline').getOrCreate()
```

```python
# read csv
df  = spark.read.csv('/content/drive/MyDrive/ML-Classification-pipeline-with-Pyspark/bank.csv', inferSchema=True, header=True)
```

```python
df.show(5)
```

```python
df.toPandas().head()
```

```python
df.dtypes
```

```python
cols = df.columns
cols
```

```python
#create list of cetgoriacal columns
catColumns= [item[0] for item in df.dtypes if item[1].startswith('string')]
catColumns
```

```python
# create list of numerical columns
numColumns = [item[0] for item in df.dtypes if item[1].startswith('int')]
numColumns
```

```python
df.select(numColumns).describe().show()
```

```python
numericData= df.select(numColumns).toPandas()
numericData.corr()
# correlation is between 0 and 1 hence we keep allthe numeric vars
```

```python
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler
```

```python
stages = []

for categoricalCol  in catColumns:
  stringIndexer = StringIndexer(inputCol= categoricalCol, outputCol= categoricalCol+ 'Index')
  encoder = OneHotEncoder(inputCols= [stringIndexer.getOutputCol()],
                          outputCols=[categoricalCol + 'classVec']
                          )
  stages += [stringIndexer, encoder]

label_strIndex = StringIndexer(inputCol= 'deposit', outputCol= 'label')
stages += [label_strIndex]

#now asemble into single feature column vector
assemblerInput = [col + "classVec" for col in catColumns] + numColumns
assembler = VectorAssembler(inputCols= assemblerInput, outputCol= "features")
stages += [assembler]
```

```python
from pyspark.ml import Pipeline
pipeline = Pipeline(stages= stages)
pipelineModel = pipeline.fit(df)
df = pipelineModel.transform(df)
selectedCols = cols + ['features', 'label']
df = df.select(selectedCols)
df.printSchema()
```

```python
df.toPandas().head(3)
```

```python
train_df, test_df = df.randomSplit([0.7, 0.3], seed = 201)
print(train_df.count(), test_df.count())
```

```python
from pyspark.ml.classification import LogisticRegression

lr = LogisticRegression(featuresCol='features', labelCol='label', maxIter=10)
lr_model = lr.fit(train_df)
```

```python
model_summary= lr_model.summary
```

```python
# get model accuracy
model_summary.accuracy
```

```python
# get precision and recall
print(model_summary.precisionByLabel)
print(model_summary.recallByLabel)
```

```python
model_summary.areaUnderROC
```

```python
roc  = model_summary.roc.toPandas()
plt.plot(roc['FPR'], roc['TPR'])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.show()
```

```python
pr  = model_summary.pr.toPandas()
plt.plot(pr['precision'], pr['recall'])
plt.xlabel('precision')
plt.ylabel('recall')
plt.title('precision vs recall')
plt.show()
```

```python
preds = lr_model.transform(test_df)

preds.select(['label', 'prediction']).toPandas().head(10)
```
